 # Menghitung ukuran pemusatan data (mean, median, modus)

```python
numerical_columns = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
```

```python
central_tendency = df[numerical_columns].agg(['mean', 'median'])
modes = df[numerical_columns].mode().iloc[0]
central_tendency.loc['mode'] = modes
print("\nMeasures of Central Tendency (Mean, Median, Mode):")
print(central_tendency)
```

# Menghitung ukuran penyebaran data (variance, std deviation, range)

```python
spread = df[numerical_columns].agg(['var', 'std'])
ranges = df[numerical_columns].max() - df[numerical_columns].min()
spread.loc['range'] = ranges

# Menghitung kuartil
quartiles = df[numerical_columns].quantile([0.25, 0.75]).T
quartiles.columns = ['Q1', 'Q3']

print("\nMeasures of Spread (Variance, Standard Deviation, Range):")
print(spread)

print("\nQuartiles (Q1, Q3):")
print(quartiles)
```

```python
# Menghitung skewness
skewness_values = df[numerical_columns].apply(skew)
print("\nSkewness of the Data:")
print(skewness_values)
```

```python
from IPython.display import display
```

```python
# Membuat plot histogram untuk setiap kolom numerik
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
axes = axes.flatten()

for i, column in enumerate(numerical_columns):
    # Menghitung mean, median, dan mode
    mean_value = df[column].mean()
    median_value = df[column].median()
    mode_value = df[column].mode()[0]

    # Membuat histogram dengan KDE
    sns.histplot(df[column].dropna(), kde=True, ax=axes[i], color='b', bins=20)
    axes[i].set_title(f'Distribusi {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frekuensi')

    # Menambahkan garis untuk mean, median, dan mode
    axes[i].axvline(mean_value, color='r', linestyle='--', label='Mean')
    axes[i].axvline(median_value, color='g', linestyle='-', label='Median')
    axes[i].axvline(mode_value, color='y', linestyle=':', label='Mode')
    
    # Menambahkan legenda
    axes[i].legend()
    
# Menghapus subplot tambahan
fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
    
# Menggabungkan hasil pemusatan, penyebaran, skewness, dan kuartil dalam satu tabel
tabel_analisis = pd.concat([central_tendency, spread], axis=0)
tabel_analisis.loc['skewness'] = skewness_values

# Menambahkan Q1 dan Q3 ke tabel utama
tabel_analisis = pd.concat([tabel_analisis, quartiles.T], axis=0)

# Menampilkan hasil dalam bentuk tabel
print("\nTabel Analisis (Pemusatan, Penyebaran, Skewness, Kuartil):")
display(tabel_analisis)
```
